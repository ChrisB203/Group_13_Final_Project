**As said, we decided to look at 3 different safety metrics**

* **Fires https://data.wprdc.org/dataset/fire-incidents-in-city-of-pittsburgh**

    **If you were going to move, you most likely wouldn't want to be going somewhere where fires break out constantly. We have gathered information on how many fires break out in each neighborhood.**
    
* **Weapon Seizure https://data.wprdc.org/dataset/pbp-fire-arm-seizures/resource/e967381d-d7e9-48e3-a2a2-39262f7fa5c4?view_id=4c4fa4d5-71cd-4aa7-92e8-18206b9140a0**

    **Moving into an area where there are constant records of illegal firearms being seized isn't the best for your safety. We compiled and ranked each neighborhood's recorded firearm seizure rate.**
    
* **Crime Rate https://data.wprdc.org/dataset/a-community-profile-of-pittsburgh-neighborhoods-1974/resource/8ce92a4b-fa62-45c3-8cee-cc58fefede75**

    **Just like the other two, something that you want to avoid. By using each neighborhood's population and how many crimes occur, we figured out each neighborhood's crime rate and ranked them.**

```python
import pandas as pd
info = pd.read_csv('WeaponSeizure.csv')
info.head(10)
```

```python
collect = info['year'] > 2020
clear = info[collect]
clear.head(10)
```

```python
graphq = clear.value_counts('neighborhood')
graphq.iloc[0:15].plot(kind = "bar")
```

```python
check = graphq > 3
final = graphq[check]
final.rank()
```
